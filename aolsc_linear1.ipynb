```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use("seaborn-poster")
```

```python
x = [14.1999, 14.2411, 14.0342, 14.2696]
y = [1981, 1983, 1985, 1987]

f = interp1d(y, x)
x_hat = f(1982)
print(x_hat)
```

```python
plt.figure(figsize=(10,8))
plt.plot(y, x, "-ob")
plt.plot(1982, x_hat, "ro")
plt.title("Temperature in 1982")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
